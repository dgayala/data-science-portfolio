3. Análisis de Correlación (2 puntos)

4. Regresión Lineal (2 puntos)
  - Implementa un modelo de regresión lineal para predecir el número de medallas obtenidas en función del número de entrenamientos semanales.
  - Obtén los coeficientes de regresión e interpreta el resultado.
  - Calcula el R² para medir el ajuste del modelo.
  - Usa Seaborn (regplot) para graficar la regresión lineal.

```python
# 4.1 Implementa el gráfico de regresión lineal usando scikit-learn
from sklearn.linear_model import LinearRegression

# Reformatear X para scikit-learn (requiere matriz de 2 dimensiones)
X = df['Entrenamientos_Semanales'].to_numpy().reshape(-1, 1) # Transforma la columna en array y luego en una matriz de n filas y 1 columna
Y = df['Medallas_Totales'] # Se queda como array

# Crear el modelo de regresión líneal vacío
modelo = LinearRegression() # Crea el modelo de la formula Y = β₀ + β₁ * X

# Ajustar (entrena) el modelo a los datos
modelo.fit(X, Y)

# Obtener coeficientes
intercepto = modelo.intercept_
pendiente = modelo.coef_[0]

# Mostrar resultados
print("\nModelo aprendido:")
print(f"Medallas obtenidas = {intercepto:.2f} + {pendiente:.2f} * Entrenamientos semanales")
print(f"\nIntercepto (β₀): {intercepto:.4f}")
print(f"Pendiente (β₁): {pendiente:.4f}")
```

```python
# 4.2.1 Obtenemos los coeficientes de regresión
intercepto = modelo.intercept_
pendiente = modelo.coef_[0]

# Mostramos  los el modelo y sus coeficientes
print(f"\nModelo aprendido: Medallas obtenidas = {intercepto:.2f} + {pendiente:.2f} * Entrenamientos semanales")
print(f"\nIntercepto (β₀): {intercepto:.4f}")
print(f"Pendiente (β₁): {pendiente:.4f}")
```

```python
# 4.2.2 Interpretación del resultado:
#
# La pendiente (β₁ = 2.154) indica que por cada entrenamiento adicional por semana,
# se espera que un atleta obtenga en promedio 2.15 medallas más, manteniendo todo lo
# demás constante. Es una relación positiva entre entrenamiento y rendimiento.
#
# El intercepto (β₀ = -7.125) representa la cantidad estimada de medallas para un atleta
# que entrena 0 veces por semana. En este contexto, no tiene un sentido práctico, ya que
# un atleta olímpico que no entrena no existiría. Es útil solo para completar el modelo matemático.
```

```python
# 4.3.1 Calculamos el R² para medir el ajuste del modelo
from sklearn.metrics import r2_score

# Obtener r cuadrado
r_cuadrado = r2_score(Y, modelo.predict(X))
print(f"\nR²: {r_cuadrado:.4f}")
```

```python
# 4.3.2 Como se ajusta el modelo
#
# El modelo explica aproximadamente un 32.5% de la variabilidad en el número
# total de medallas a partir del número de entrenamientos semanales.
# Es un valor moderado, lo que indica que: Aunque el entrenamiento semanal influye
# en el rendimiento, otros factores también juegan un rol importante, como disciplina,
# técnica, motivación o condiciones físicas.
```

```python
# 4.4 Usamos regplot() para graficar la regresión lineal
plt.figure(figsize=(6, 4))
sns.regplot(x='Entrenamientos_Semanales', y='Medallas_Totales', data=df, ci=None)
plt.title('Regresión Lineal: Entrenamientos vs Medallas')
plt.xlabel('Entrenamientos por Semana')
plt.ylabel('Medallas Totales')
plt.grid(True, linestyle='--', linewidth='0.6', zorder=1)
plt.show()
```

5. Visualización de Datos con Seaborn y Matplotlib (2 puntos)
  - Crea un heatmap de correlación entre todas las variables numéricas.
  - Crea un boxplot de la cantidad de medallas por disciplina deportiva.
  - Personaliza los gráficos con títulos, etiquetas y colores.

```python
# 5.1 Creamos un heatmap de correlación entre todas las variables numéricas
# Filtramos solo columnas numéricas
df_numericas = df.select_dtypes(include='number')

# Calculamos la matriz de correlación
correlaciones = df_numericas.corr()

# Graficamos el heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlaciones, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor de Correlación entre Variables Numéricas')
plt.tight_layout()
plt.show()
```

```python
# 5.2 Creamos un boxplot de la cantidad de medallas por disciplina deportiva
plt.figure(figsize=(10, 6))
sns.boxplot(x='Deporte', y='Medallas_Totales', data=df)
plt.show()
```

```python
# 5.3 Personalizamos los gráficos con títulos, etiquetas y colores
plt.figure(figsize=(10, 6))
sns.boxplot(x='Deporte', y='Medallas_Totales', hue='Deporte', data=df,
            palette='Set2', dodge=False)
plt.title('Distribución de Medallas por Disciplina Deportiva')
plt.xlabel('Disciplina')
plt.ylabel('Medallas Totales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
